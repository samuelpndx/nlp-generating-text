````python
# Initialize the vectorization layer
vectorize_layer = tf.keras.layers.TextVectorization()

# Build the vocab
vectorize_layer.adapt(corpus)
````

````python
# Ge the vocab
vocabulary = vectorize_layer.get_vocabulary()

# Get the vocab size
vocab_size = len(vocabulary)

print(f'Vocabulary size: {vocab_size}')
print(vocabulary)
````

# Preprocessing the Dataset

Each line of the song create inputs and labels from it. For example, if you only have one sentence: "Vou pedir licença pra contar a minha historia", you want the model to learn the next word given any subphrase of this sentence:

```
INPUT                                        LABEL
-----------------------------------------------------
Vou                                     ---> pedir
Vou pedir                               ---> licença
Vou pedir licença                       ---> pra
Vou pedir licença pra                   ---> contar
Vou pedir licença pra contar            ---> a
Vou pedir licença pra contar a          ---> minha
Vou pedir licença pra contar a minha    ---> historia
```

````python
input_sequences = []

for line in corpus:
    sequence = vectorize_layer(line).numpy()
    for i in range(1,len(sequence)):
        n_gram_sequence = sequence[:i+1]
        input_sequences.append(n_gram_sequence)

max_sequence_len = max([len(x) for x in input_sequences])

input_sequences = np.array(tf.keras.utils.pad_sequences(input_sequences,
                                                        maxlen=max_sequence_len,
                                                        padding='pre'))

xs, labels = input_sequences[:,:-1], input_sequences[:,-1]

ys = tf.keras.utils.to_categorical(labels, num_classes=vocab_size)
````

````python
def sequence_to_text(sequence, vocabulary):
    words = [vocabulary[index] for index in sequence]

    text = tf.strings.reduce_join(words, separator=' ').numpy().decode()

    return text
````

# Build the model

````python
model = tf.keras.Sequential([
    tf.keras.Input(shape=(max_sequence_len-1,)),
    tf.keras.layers.Embedding(vocab_size, 64),
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(20)),
    tf.keras.layers.Dense(vocab_size, activation='softmax')
])

model.compile(
    loss='categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

model.summary()
````

# Train the model

````python
history = model.fit(xs, ys, epochs=500)
````

````python
import matplotlib.pyplot as plt

# Plot utility
def plot_graphs(history, string):
  plt.plot(history.history[string])
  plt.xlabel("Epochs")
  plt.ylabel(string)
  plt.show()

# Visualize the accuracy
plot_graphs(history, 'accuracy')
````

# Generating text

1. Feed a seed text to initiate the process.
2. Model predicts the index of the most probable next word.
3. Look up the index in the reverse word index dictionary
4. Append the next word to the seed text.
5. Feed the result to the model again.

````python
seed_text = "Como um vaqueiro tem suas perdas e suas glorias"

next_words = 100

for _ in range(next_words):
    sequence = vectorize_layer(seed_text)

    sequence = tf.keras.utils.pad_sequences([sequence],
                                             maxlen=max_sequence_len-1,
                                             padding='pre')

    probabilities = model.predict(sequence, verbose=0)

    predicted = np.argmax(probabilities, axis=-1)[0]

    if predicted != 0:
        output_word = vocabulary[predicted]
        seed_text += " " + output_word

print(seed_text)
````

````python
seed_text = "Como um vaqueiro tem suas perdas e suas glórias"

next_word = 100

for _ in range(next_word):
    sequence = vectorize_layer(seed_text)
    sequence = tf.keras.utils.pad_sequences([sequence],
                                            maxlen=max_sequence_len-1,
                                            padding='pre')
    probabilities = model.predict(sequence, verbose=0)

    choice = np.random.choice([1,2,3])

    predicted = np.argsort(probabilities)[0][-choice]

    if predicted != 0:
        output_word = vocabulary[predicted]
        seed_text += " " + output_word

print(seed_text)
````
